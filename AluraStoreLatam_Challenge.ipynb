### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
tienda1_total_ingresos = tienda['Precio'].sum()
print(f'El ingreso total de la primera tienda es: ${tienda1_total_ingresos}')
tienda2_total_ingresos = tienda2['Precio'].sum()
print(f'El ingreso total de la segunda tienda es: ${tienda2_total_ingresos}')
tienda3_total_ingresos = tienda3['Precio'].sum()
print(f'El ingreso total de la tercera tienda es: ${tienda3_total_ingresos}')
tienda4_total_ingresos = tienda4['Precio'].sum()
print(f'El ingreso total de la cuarta tienda es: ${tienda4_total_ingresos}')
```

# 2. Ventas por categoría

```python
Tienda1_categoriaVentas= tienda.groupby(['Categoría del Producto'])['Categoría del Producto'].count().sort_values(ascending=False)
Tienda2_categoriaVentas= tienda2.groupby(['Categoría del Producto'])['Categoría del Producto'].count().sort_values(ascending=False)
Tienda3_categoriaVentas= tienda3.groupby(['Categoría del Producto'])['Categoría del Producto'].count().sort_values(ascending=False)
Tienda4_categoriaVentas= tienda4.groupby(['Categoría del Producto'])['Categoría del Producto'].count().sort_values(ascending=False)
```

```python
ventas_por_categoria = pd.DataFrame({
    'Tienda 1': Tienda1_categoriaVentas,
    'Tienda 2': Tienda2_categoriaVentas,
    'Tienda 3': Tienda3_categoriaVentas,
    'Tienda 4': Tienda4_categoriaVentas
})
ventas_categoria_mayormenor = ventas_por_categoria.sort_values(by='Tienda 1', ascending=False)
ventas_categoria_mayormenor
```

### **Visualizacion grafica sobre la popularidad de cada producto por tienda**

de mayor a menor

```python
ventas_categoria_mayormenor.plot(kind='bar', figsize=(14, 8), xlabel='Categoría del Producto', ylabel='Número de Ventas')
```

sin filtro

```python
ventas_por_categoria.plot(kind='bar', figsize=(14, 8), xlabel='Categoría del Producto', ylabel='Número de Ventas')
```

### Resumen de Hallazgos

Analizando las ventas por categoría de producto en las cuatro tiendas (`ventas_consolidadas`) y la visualización generada:

*   **Categorías más vendidas:** Los `Muebles` y `Electrónicos` consistentemente muestran el mayor número de ventas en todas las tiendas.
    *   `Muebles` lidera en Tienda 1 (465), Tienda 3 (499) y Tienda 4 (480).
    *   `Electrónicos` lidera en Tienda 2 (422), y es muy fuerte en el resto.
*   **Categorías menos vendidas:** `Libros`, `Artículos para el hogar`, e `Instrumentos musicales` son las categorías con las ventas más bajas en todas las tiendas.
*   **Variación entre tiendas:**
    *   La **Tienda 3** tiene la mayor venta de `Muebles` (499 unidades).
    *   La **Tienda 4** muestra la mayor venta de `Artículos para el hogar` (201 unidades) y `Juguetes` (338 unidades).
    *   La **Tienda 2** tiene la mayor venta de `Instrumentos musicales` (224 unidades) y `Libros` (197 unidades).
    *   `Electrodomésticos` tiene una variabilidad notable, siendo la más alta en Tienda 1 (312) y la más baja en Tienda 4 (254).

En general, las tiendas exhiben patrones de venta similares en las categorías principales, con algunas variaciones en las categorías de nicho, lo que sugiere posibles especializaciones o preferencias de clientes por tienda.
