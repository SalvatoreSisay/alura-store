# 5. Envío promedio por tienda

```python
def costo_promedio(nombre, tienda):
    costo_envio = tienda['Costo de envío'].mean()
    print(f'El costo de envío promedio de la {nombre} es: ${round(costo_envio,2)}')
    return costo_envio
```

```python
costo_promedio('primera tienda', tienda)
costo_promedio('segunda tienda', tienda2)
costo_promedio('tercera tienda', tienda3)
costo_promedio('cuarta tienda', tienda4)
```

```python
nombres_tiendas_lista = list(nombre_tiendas)
costo_envio_t1 = costo_promedio('primera tienda', tienda)
costo_envio_t2 = costo_promedio('segunda tienda', tienda2)
costo_envio_t3 = costo_promedio('tercera tienda', tienda3)
costo_envio_t4 = costo_promedio('cuarta tienda', tienda4)

costos_promedio_lista = [costo_envio_t1,costo_envio_t2,costo_envio_t3,costo_envio_t4]

print("Nombres de las tiendas:", nombres_tiendas_lista)
print("Costos promedio de envío:", costos_promedio_lista)
```

```python
plt.figure(figsize=(12, 8))
plt.plot(nombres_tiendas_lista, costos_promedio_lista, marker = 'o', linestyle = '-', color='blue')
plt.title('Costos de Envío Promedio por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Costo de Envío Promedio')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
