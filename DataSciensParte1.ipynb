## Calcular ventas por producto

### Subtask:
Calculate the sales for each unique product in each of the four stores to determine their individual popularity.


**Reasoning**:
To calculate the sales for each unique product in each store, I need to group each DataFrame by the 'Producto' column and count the occurrences of each product, storing the results in new variables as specified.



```python
ventas_por_producto_tienda1 = tienda.groupby('Producto')['Producto'].count().sort_values(ascending=False)
ventas_por_producto_tienda2 = tienda2.groupby('Producto')['Producto'].count().sort_values(ascending=False)
ventas_por_producto_tienda3 = tienda3.groupby('Producto')['Producto'].count().sort_values(ascending=False)
ventas_por_producto_tienda4 = tienda4.groupby('Producto')['Producto'].count().sort_values(ascending=False)

print("Ventas por producto en Tienda 1:")
print(ventas_por_producto_tienda1.head())
print("\nVentas por producto en Tienda 2:")
print(ventas_por_producto_tienda2.head())
print("\nVentas por producto en Tienda 3:")
print(ventas_por_producto_tienda3.head())
print("\nVentas por producto en Tienda 4:")
print(ventas_por_producto_tienda4.head())
```

## Identificar productos más y menos vendidos

### Subtask:
Identificar los productos con el mayor y menor número de ventas en cada tienda.


**Reasoning**:
To identify the top 5 and bottom 5 selling products for each store, I will use the `.head(5)` method on the already sorted product sales series for the top products and `.tail(5)` for the bottom products, and store them in separate variables.



```python
top5_tienda1 = ventas_por_producto_tienda1.head(5)
top5_tienda2 = ventas_por_producto_tienda2.head(5)
top5_tienda3 = ventas_por_producto_tienda3.head(5)
top5_tienda4 = ventas_por_producto_tienda4.head(5)

bottom5_tienda1 = ventas_por_producto_tienda1.tail(5)
bottom5_tienda2 = ventas_por_producto_tienda2.tail(5)
bottom5_tienda3 = ventas_por_producto_tienda3.tail(5)
bottom5_tienda4 = ventas_por_producto_tienda4.tail(5)

print("Top 5 productos más vendidos en Tienda 1:\n", top5_tienda1)
print("\nBottom 5 productos menos vendidos en Tienda 1:\n", bottom5_tienda1)

print("\nTop 5 productos más vendidos en Tienda 2:\n", top5_tienda2)
print("\nBottom 5 productos menos vendidos en Tienda 2:\n", bottom5_tienda2)

print("\nTop 5 productos más vendidos en Tienda 3:\n", top5_tienda3)
print("\nBottom 5 productos menos vendidos en Tienda 3:\n", bottom5_tienda3)

print("\nTop 5 productos más vendidos en Tienda 4:\n", top5_tienda4)
print("\nBottom 5 productos menos vendidos en Tienda 4:\n", bottom5_tienda4)
```

## Consolidar y preparar datos para visualización

### Subtask:
Organizar los productos más y menos vendidos de todas las tiendas en una estructura de datos que facilite su comparación y visualización conjunta.


**Reasoning**:
First, I will create dictionaries for the top 5 and bottom 5 selling products for each store to prepare them for DataFrame conversion, as specified in the instructions.



```python
top_products_dict = {
    'Tienda 1': top5_tienda1,
    'Tienda 2': top5_tienda2,
    'Tienda 3': top5_tienda3,
    'Tienda 4': top5_tienda4
}

bottom_products_dict = {
    'Tienda 1': bottom5_tienda1,
    'Tienda 2': bottom5_tienda2,
    'Tienda 3': bottom5_tienda3,
    'Tienda 4': bottom5_tienda4
}

print("Top Products Dictionary:")
for store, products in top_products_dict.items():
    print(f"{store}:\n{products.head(2)}")

print("\nBottom Products Dictionary:")
for store, products in bottom_products_dict.items():
    print(f"{store}:\n{products.head(2)}")
```

**Reasoning**:
To organize the top and bottom selling products for comparison, I will convert the dictionaries created in the previous step into DataFrames, reshape them into a long format, add a 'Tipo de Producto' column to distinguish between top and bottom sellers, and then concatenate these two DataFrames.



```python
import pandas as pd

df_top_products = pd.DataFrame.from_dict(top_products_dict)
df_top_products = df_top_products.stack().reset_index()
df_top_products.columns = ['Producto', 'Tienda', 'Ventas']
df_top_products['Tipo de Producto'] = 'Más Vendidos'

df_bottom_products = pd.DataFrame.from_dict(bottom_products_dict)
df_bottom_products = df_bottom_products.stack().reset_index()
df_bottom_products.columns = ['Producto', 'Tienda', 'Ventas']
df_bottom_products['Tipo de Producto'] = 'Menos Vendidos'

df_consolidated_products = pd.concat([df_top_products, df_bottom_products])

print("Consolidated DataFrame of Top and Bottom Products:")
print(df_consolidated_products.head(10))
```

## Visualizar productos destacados

### Subtask:

**Reasoning**:
To visualize the top and bottom selling products for each store as requested, I need to import `seaborn` and `matplotlib.pyplot`, and then use `seaborn.catplot` with appropriate parameters for faceting, x and y axes, and labels.



```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(18, 10))

g = sns.catplot(
    data=df_consolidated_products,
    x='Producto',
    y='Ventas',
    hue='Tienda',
    col='Tipo de Producto',
    kind='bar',
    col_wrap=2, # Display 2 columns for the facets
    height=6, aspect=1.5, # Adjust figure size for better readability
    sharey=False # Allow y-axis to scale independently for better comparison
)

# Rotate x-axis labels for readability for each subplot
for ax in g.axes.flat:
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')

g.set_axis_labels("Producto", "Número de Ventas")
g.set_titles("{col_name}")
g.fig.suptitle('Productos Más y Menos Vendidos por Tienda', y=1.02) # Set overall title

plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```
